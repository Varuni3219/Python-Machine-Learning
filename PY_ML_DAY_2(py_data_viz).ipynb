##NUMPY


####Creating Arrays

```python
import numpy as np
a=np.array([1,2,3,4,5])
b=np.arange(1,6) # Creates[1,2,3,4,5]
c=np.zeros((2,3)) #2x3 matrix of zeros
d=np.ones((3,2)) #3x2 matrix of ones

print("Array a:",a)
print("Arrange b:",b)
print("2x3 Zeros:",c)
print("3x2 Ones:",d)
```

####Indexing & Silicing

```python
print("First Element of a:",a[0])
print("Last 2 elements of a:",a[-2:])
```

####Basic Arithmetic Operations

```python
print("a+2:",a+2)
print("a*3:",a*3)
```

####Basic Statistics

```python
print("Square of a:",np.square(a))
print("Mean of a:",np.mean(a))
print("Standard Deviation of a:",np.std(a))
```

####Reshaping Arrays

```python
matrix=np.arange(1,10).reshape((3,3))
print("Reshaped 3x3 matrix:")
matrix
```

####Transpose

```python
print("Transpose:")
matrix.T
```

####Boolean Indexing

```python
print("Elements > 5:",matrix[matrix > 5])
```

##**Pandas**

####Create a Data Frame Manually

```python
import pandas as pd

data={
    'Name':['Alice','Bob','Charlie','David','Eva'],
    'Age':[25,30,35,40,22],
    'Department':['Sales','Engineering','HR','Engineering','Sales'],
    'Salary':[50000,80000,40000,90000,45000]
}
df=pd.DataFrame(data)
df
```

####Basic Operations


```python
print("First 3 rows:")
df.head(3)
```

```python
print("Column names: ")
df.columns
```

```python
print("Data Types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

####Filtering

```python
print("People with salary > 50000:")
df[df['Salary']>50000]
```

####Sorting

```python
print("Sorted by age:")
df.sort_values(by='Age')
```

####Grouping & Aggregation

```python
grouped=df.groupby('Department')['Salary'].mean()
print("Avearge salary by Department:")
grouped
```

####Adding new columns

```python
df['Salary in Lakhs']=df['Salary']/100000
df
```

##**MatPlotLib: Building Block Of Visualization**

####Line Plot

```python
import matplotlib.pyplot as plt

years=[2019,2020,2021,2022]
revenue=[100,120,130,160]

plt.plot(years,revenue,marker='o',linestyle='--',color='red')
plt.title("Company Revenue Over Years")
plt.xlabel("Year")
plt.ylabel("Revenue (in crores)")
plt.xticks(years)
plt.grid(True)
plt.show()
```

####Bar Chart

```python
departments=df['Department'].value_counts()
departments.plot(kind='bar',color='lavender')
plt.title("Employment Count By Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

####Histogram

```python
ages=df['Age']
plt.hist(ages,bins=5,color='blue',edgecolor='indigo')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

####Pie Chart

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```

##**SEABORN:Statistical Data Visulization**

####Load Sample Data Set

```python
import seaborn as sns

tips=sns.load_dataset("tips")
tips.head()
```

```python
print("\n Info about the tips dataset:")
tips.info()

print("\n Description of the tips dataset:")
tips.describe(include='all')
```

####Pair Plot

```python
sns.pairplot(tips,hue="sex")
plt.show()
```

####Heatmap Of Correlation

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Martix")
plt.show()
```

###Count Plot

```python
sns.countplot(x='day',data=tips,palette='Set2',hue='day',legend=False)
plt.title("Count Of Records By Day")
plt.show()
```

####Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box Plot Of Total Bill By Day")
plt.show()
```
